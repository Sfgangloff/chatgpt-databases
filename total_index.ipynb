```python
import yaml
```

```python
with open('data/molecular_biology/with_index_w.yaml', 'r') as file:
    yaml_content = yaml.safe_load(file)
```

```python
print(yaml_content["sections"])
total_index = []
sections = []
for section in yaml_content["sections"]:
    for subsection in section["subsections"]:
        for subsubsection in subsection["subsubsections"]:
            sections.append(section["title"] + "-" + subsection["title"] + "-" + subsubsection["title"])
            total_index.extend([word["word"] for word in subsubsection["index"]["words"]])
```

```python
total_index = list(set(total_index))
```

```python
for word in total_index:
    print(word)
```

```python
for section in sections:
    print(section)
```
